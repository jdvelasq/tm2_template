```python
import pickle

from techminer2.science_mapping.co_occurrence import communities_summary

with open('reports/custom_items.pkl', 'rb') as file:
    custom_items = pickle.load(file)

m = communities_summary(
    field="descriptors",
    conserve_counters=False,
    top_n=None,
    occ_range=(None, None),
    custom_items=custom_items,
    algorithm_or_dict="louvain",
    association_index="association",
    root_dir="./",
    year_filter=(2014, 2023),
)

m["Terms"] = m["Terms"].str.split("; ")
m["Terms"] = m["Terms"].map(lambda x: [t.strip().replace("_", " ").title() for t in x[:10]])
m["Terms"] = m["Terms"].str.join("; ")

m.to_csv("reports/clusters_summary.csv", sep="\t")

print(m.to_markdown())
```
