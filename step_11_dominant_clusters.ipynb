```python
from techminer2.analyze.co_occurrence.network import to_brute_force_labels
from techminer2.analyze.pca.cooc_matrix.brute_force import factor_map
```

```python
#
# Generate a dictionary with the filtered terms (by time period and
# occurrence) and the assigned cluster
brute_force_labels = to_brute_force_labels(
    #
    # PARAMS:
    field="author_keywords",
    #
    # COLUMN PARAMS:
    top_n=None,
    occ_range=(3, None),
    gc_range=(None, None),
    custom_items=None,
    #
    # NETWORK PARAMS:
    algorithm_or_dict="louvain",
    association_index="association",
    #
    # DATABASE PARAMS:
    root_dir="./",
    database="main",
    year_filter=(2014, 2023),
    cited_by_filter=(None, None),
)

#
# Generate a list of custom items conformed by the author keywords
# used to obtain the clusters
custom_items = [" ".join(term.split(" ")[:-1]) for term in brute_force_labels.keys()]
```

```python
factor_map(
    #
    # PARAMS:
    field="author_keywords",
    association_index=None,
    #
    # ITEM PARAMS:
    top_n=None,
    occ_range=(3, None),
    gc_range=(None, None),
    custom_items=custom_items,
    #
    # PCA PARAMS:
    n_components=5,
    whiten=False,
    svd_solver="auto",
    pca_tol=0.0,
    iterated_power="auto",
    n_oversamples=10,
    power_iteration_normalizer="auto",
    random_state=0,
    #
    # BRUTE FORCE PARAMS:
    brute_force_labels=brute_force_labels,
    #
    # LAYOUT:
    nx_k=None,
    nx_iterations=5,
    nx_random_state=4,
    #
    # NODES:
    node_color="#7793a5",
    node_size_range=(16, 40),
    textfont_size_range=(8, 13),
    textfont_opacity_range=(0.35, 1.00),
    #
    # EDGES:
    edge_top_n=None,
    edge_similarity_min=None,
    edge_widths=(2, 2, 3, 3),
    edge_colors=("#7793a5", "#7793a5", "#7793a5", "#7793a5"),
    #
    # AXES:
    xaxes_range=None,
    yaxes_range=None,
    show_axes=False,
    #
    # DATABASE PARAMS:
    root_dir="./",
    database="main",
    year_filter=(2014, 2023),
    cited_by_filter=(None, None),
).update_layout(width=900, height=700)
```
